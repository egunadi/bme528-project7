## compare Intcornea (the output of OCT_InnerCornea()) from both the .mat file and the debug output

```python
mat_intcornea = mat_data['Intcornea']
py_intcornea = debug_vars['Intcornea']

print("MATLAB Intcornea fields:", mat_intcornea.dtype.names)
print("Python Intcornea fields:", py_intcornea.keys())
```

```python
def compare_field(field):
    mat_val = mat_intcornea[field][0, 0]
    py_val = py_intcornea[field]

    print(f"\n--- Field: {field} ---")
    if isinstance(mat_val, np.ndarray) and isinstance(py_val, np.ndarray):
        print(f"MATLAB shape: {mat_val.shape}, min: {np.min(mat_val)}, max: {np.max(mat_val)}, mean: {np.mean(mat_val)}")
        print(f"Python shape: {py_val.shape}, min: {np.min(py_val)}, max: {np.max(py_val)}, mean: {np.mean(py_val)}")

        if mat_val.shape == py_val.shape:
            diff = np.abs(mat_val.astype(np.float32) - py_val.astype(np.float32))
            print(f"Diff — min: {np.min(diff)}, max: {np.max(diff)}, mean: {np.mean(diff)}")
        else:
            print("⚠️ Shape mismatch — skipping diff")
    else:
        print(f"MATLAB value: {mat_val}")
        print(f"Python value: {py_val}")

# Then run:
for field in py_intcornea:
    if field in mat_intcornea.dtype.names:
        compare_field(field)
```

## compare inputs and outputs of OuterDewarp() from both the .mat file and the debug output

### input comparison

```python
# Image stats
def print_image_stats(name, img):
    print(f"{name} — shape: {img.shape}, dtype: {img.dtype}, min: {img.min()}, max: {img.max()}, mean: {img.mean()}")

print_image_stats("MATLAB im_s", mat_data['im_s'])
print_image_stats("Python im_s", debug_vars['im_s'])

print_image_stats("MATLAB im_t", mat_data['im_t'])
print_image_stats("Python im_t", debug_vars['im_t'])

# Numeric parameters
param_names = ['D', 'w', 'd', 'n_tissue1', 'x_dimension', 'y_dimension', 'ShowColors']
for param in param_names:
    mat_val = mat_data[param]
    py_val = debug_vars[param]
    print(f"{param} — MATLAB: {mat_val} | Python: {py_val}")

# Spline check
# Step 1: Unpack the MATLAB PPout struct
mat_ppout_struct = mat_data['PPout'][0, 0]  # it's a tuple

# Step 2: The 3rd element is 'coefs' (index 2)
mat_ppout_coeffs = mat_ppout_struct[2]  # this is a (40, 4) array

# Step 3: Python side — already stored in debug_vars['PPout']
py_ppout_coeffs = debug_vars['PPout']
py_ppout_coeffs = py_ppout_coeffs.T  # transpose to (40, 4)

# Step 4: Print and compare
print("MATLAB PPout coeffs — shape:", mat_ppout_coeffs.shape)
print("Python PPout coeffs — shape:", py_ppout_coeffs.shape)

min_len = min(mat_ppout_coeffs.shape[0], py_ppout_coeffs.shape[0])
diff = np.abs(mat_ppout_coeffs[:min_len] - py_ppout_coeffs[:min_len])
print("PPout diff — mean:", diff.mean())

if abs(mat_ppout_coeffs.shape[0] - py_ppout_coeffs.shape[0]) <= 6:
    print("✅ PPout spline segment count is close enough — difference is negligible.")
else:
    print(f"⚠️ PPout spline shapes differ noticeably — MATLAB: {mat_ppout_coeffs.shape[0]}, Python: {py_ppout_coeffs.shape[0]}")
```

## compare inputs and outputs of InnerDewarp() from both the .mat file and the debug output

### input comparison

```python
# Image stats
print_image_stats("MATLAB im_s", mat_data['im_s'])
print_image_stats("Python im_s", debug_vars['im_s'])

print_image_stats("MATLAB dewarpedOut", mat_data['dewarpedOut'])
print_image_stats("Python dewarpedOut", debug_vars['dewarpedOut'])

# Numeric parameters
param_names = ['D', 'w', 'd', 'n_tissue1', 'n_tissue2', 'x_dimension', 'y_dimension', 'ShowColors']
for param in param_names:
    print(f"{param} — MATLAB: {mat_data[param]} | Python: {debug_vars[param]}")

# Spline coefficients (PPout and PPinn)
mat_ppout = mat_data['PPout'][0, 0][2]  # [2] = 'coefs'
py_ppout = debug_vars['PPout'].T  # transpose for shape match

mat_ppinn = mat_data['PPinn'][0, 0][2]
py_ppinn = debug_vars['PPinn'].T

# Compare shapes and diffs
print("PPout shape comparison:", mat_ppout.shape, py_ppout.shape)
print("PPinn shape comparison:", mat_ppinn.shape, py_ppinn.shape)

min_len = min(mat_ppout.shape[0], py_ppout.shape[0])
diff = np.abs(mat_ppout[:min_len] - py_ppout[:min_len])
print("PPout diff — mean:", diff.mean())
if abs(mat_ppinn.shape[0] - py_ppinn.shape[0]) <= 6:
    print("✅ Spline segment counts are close enough — difference is negligible.")
else:
    print("⚠️ PPinn spline shapes differ noticeably — consider checking the input x-values.")
```

### output comparison

```python
# Image stats
def print_image_stats(name, img):
    print(f"{name} — shape: {img.shape}, dtype: {img.dtype}, min: {img.min()}, max: {img.max()}, mean: {img.mean()}")

# Coordinate stats
def print_coords_stats(name, coords):
    print(f"{name} — shape: {coords.shape}, min: {np.min(coords)}, max: {np.max(coords)}, mean: {np.mean(coords)}")

# Image
print_image_stats("MATLAB dewarpedFull", mat_data['dewarpedFull'])
print_image_stats("Python dewarpedFull", debug_vars['dewarpedFull'])

# Coordinates
print_coords_stats("MATLAB x_s2", mat_data['x_s2'])
print_coords_stats("Python x_s2", debug_vars['x_s2'])

print_coords_stats("MATLAB y_s2", mat_data['y_s2'])
print_coords_stats("Python y_s2", debug_vars['y_s2'])
```
