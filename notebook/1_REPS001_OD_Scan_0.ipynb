## compare Extcornea (the output of OCT_OuterCornea() and input of OCT_InnerCornea()) from both the .mat file and the debug output

```python
print("MATLAB Extcornea fields:", mat_data['Extcornea'].dtype)
print("Python Extcornea fields:", debug_vars['Extcornea'].keys())
```

```python
import numpy as np

for field in debug_vars['Extcornea'].keys():
    print(f"\n--- Field: {field} ---")
    mat_val = mat_data['Extcornea'][field][0, 0]  # MATLAB .mat struct access
    py_val = debug_vars['Extcornea'][field]

    # Handle arrays
    if isinstance(mat_val, np.ndarray) and isinstance(py_val, np.ndarray):
        print("MATLAB shape:", mat_val.shape, "min:", np.min(mat_val), "max:", np.max(mat_val), "mean:", np.mean(mat_val))
        print("Python shape:", py_val.shape, "min:", np.min(py_val), "max:", np.max(py_val), "mean:", np.mean(py_val))

        # Optional: compute difference
        if mat_val.shape == py_val.shape:
            diff = np.abs(mat_val.astype(np.float64) - py_val.astype(np.float64))
            print("Diff  min:", np.min(diff), "max:", np.max(diff), "mean:", np.mean(diff))
        else:
            print(" Shape mismatch  skipping diff")
    else:
        print("MATLAB value:", mat_val)
        print("Python value:", py_val)
```

## compare Intcornea (the output of OCT_InnerCornea()) from both the .mat file and the debug output

```python
mat_intcornea = mat_data['Intcornea']
py_intcornea = debug_vars['Intcornea']

print("MATLAB Intcornea fields:", mat_intcornea.dtype.names)
print("Python Intcornea fields:", py_intcornea.keys())
```

```python
def compare_field(field):
    mat_val = mat_intcornea[field][0, 0]
    py_val = py_intcornea[field]

    print(f"\n--- Field: {field} ---")
    if isinstance(mat_val, np.ndarray) and isinstance(py_val, np.ndarray):
        print(f"MATLAB shape: {mat_val.shape}, min: {np.min(mat_val)}, max: {np.max(mat_val)}, mean: {np.mean(mat_val)}")
        print(f"Python shape: {py_val.shape}, min: {np.min(py_val)}, max: {np.max(py_val)}, mean: {np.mean(py_val)}")

        if mat_val.shape == py_val.shape:
            diff = np.abs(mat_val.astype(np.float32) - py_val.astype(np.float32))
            print(f"Diff  min: {np.min(diff)}, max: {np.max(diff)}, mean: {np.mean(diff)}")
        else:
            print(" Shape mismatch  skipping diff")
    else:
        print(f"MATLAB value: {mat_val}")
        print(f"Python value: {py_val}")

# Then run:
for field in py_intcornea:
    if field in mat_intcornea.dtype.names:
        compare_field(field)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Safely extract and flatten the MATLAB xcornea and ycornea
mat_x_raw = mat_data['Intcornea']['xcornea']
mat_y_raw = mat_data['Intcornea']['ycornea']

# Unwrap if nested inside an object array
if isinstance(mat_x_raw[0, 0], np.ndarray):
    mat_x = mat_x_raw[0, 0].flatten()
    mat_y = mat_y_raw[0, 0].flatten()
else:
    mat_x = mat_x_raw.flatten()
    mat_y = mat_y_raw.flatten()

# Python debug output
py_x = debug_vars['Intcornea']['xcornea']
py_y = debug_vars['Intcornea']['ycornea']

# Plot it!
plt.figure(figsize=(10, 6))
plt.imshow(debug_vars['Extcornea']['originalgray'], cmap='gray')
plt.plot(mat_x, mat_y, 'r-', label='MATLAB')
plt.plot(py_x, py_y, 'g--', label='Python')
plt.title("Inner Cornea Boundary: MATLAB vs Python")
plt.legend()
plt.show()
```
